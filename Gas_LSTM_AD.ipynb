# LSTM Model

```python
timesteps = 12
data_dim = 60
output_dim=60
```

```python
xy = np.asarray(df2)
```

```python
xy
```

```python
train_size= int(len(xy)*0.7)
train_set = xy[:train_size]
test_set = xy[train_size-timesteps:]
```

```python
print(test_set)
```

```python
def build_dataset(time_series, timesteps):
    dataX = []
    dataY = []
    for i in range(0,len(time_series) - timesteps):
        _x = time_series[i:i + timesteps, :]
        _y = time_series[i + timesteps, :]
        print(_x, "->", _y)
        dataX.append(_x)
        dataY.append(_y)
    return dataX, dataY
```

```python
trainX, trainY = build_dataset(train_set, timesteps)
testX, testY = build_dataset(test_set, timesteps)
```

```python
# input place holders
X = tf.placeholder(tf.float32, [None, 12, 60])
Y = tf.placeholder(tf.float32, [None, 60])
```

```python
## Resetting Graph
tf.reset_default_graph()
```

```python
# build a LSTM network

hidden_dim = 10
learning_rate = 0.01
iterations = 500

basic_cell = tf.nn.rnn_cell.LSTMCell(name='basic_lstm_cell', num_units=hidden_dim, state_is_tuple=True) ## state_is_tuple?

outputs, states = tf.nn.dynamic_rnn(basic_cell, X, dtype=tf.float32) 
## Cell에 입력 X, 입력 데이터형태 제시,
## (matmul() 등을 실행하여 계속 가중치 W, 편향 b들을 업데이트)
## static_rnn 함수의 결과값은 두 가지 객체임. 1. 각 타임스텝에서의 출력 텐서를 담고 있는 파이썬 리스트 2. 네트워크의 최종 상태 텐서

Y_pred = tf.contrib.layers.fully_connected(
    outputs, output_dim, activation_fn=None)  # We use the last cell's output
```

```python
# Loss Function

loss = tf.reduce_sum(tf.square(Y_pred-Y))

#Optimizer

optimizer = tf.train.AdamOptimizer(learning_rate)
train = optimizer.minimize(loss)

# RMSE

targets = tf.placeholder(tf.float32, [None, 1])
predictions = tf.placeholder(tf.float32, [None, 1])
rmse = tf.sqrt(tf.reduce_mean(tf.square(targets-predictions)))
```

```python
with tf.Session() as sess:
    init = tf.global_variables_initializer()
    sess.run(init)
    
    #Training Steps
    
    for i in range(iterations):
        _,step_loss = sess.run([train,loss], feed_dict={X: trainX, Y:trainY})
        print("[Step: {}] loss: {}".format(i, step_loss))
        
    test_predict = sess.run(Y_pred, feed_dict={X: testX})
    rmse_val = sess.run(rmse, feed_dict={targets: testY, predictions: test_predict})
    print("RMSE: {}".format(rmse_val))
    
```
